Ways to create the DataFrame
1. from lists. 1D or 2D Lists
2. from np Arrays. 1D or 2D array. Set the Index and Column
3. from Dictionaries
4. From an existing DF. Select columns using [[]] which selects the specific series and makes them into a DF

```python
import numpy as np
import pandas as pd
```

```python
# create using 2D lists
listRand = [[14, 20, 50], [50, 60, 90]]
listRand
dfRand = pd.DataFrame(listRand, index=["a", "b"], columns=["A", "B", "C"])
dfRand
```

```python
#create a DF using dict containing numpy arrays
arrCars = np.array(["Accent", "Scorpio", "XUV", "Highlander", "Sonata", "Tesla"])
arrYearBought = np.array([2001, 2005, 2012, 2022, 2022, 2025])
arrColor = np.array(["Black", "Beige", "White", "White", "Blue", "Red"])

dfCars = pd.DataFrame({"Year":arrYearBought, "Color":arrColor}, arrCars)
dfCars
```

```python
dfCars.columns = ["Y", "C"]
dfCars
```

```python
np.random.seed(42)
arr2D = np.random.randint(0, 100, (5, 4))
arr2D
```

```python
df2 = pd.DataFrame(arr2D, columns=["C1", "C2", "C3", "C4"], index=["R0","R1","R2","R3", "R4"])
df2
```

```python
dfCars["Y"][1:5]
```

```python
np.random.seed(42)
df3 = pd.DataFrame(np.random.rand(6, 5), index=["A", "B", "C", "D", "E", "F"], columns=["V1", "V2", "V3", "V4", "V5"])
df3
```

```python
#pull out a specific column as a series
df3["V1"]
```

```python
# pull out 2 columns as a DF
df3[["V2", "V5"]]

#pull out row as a DF
df3["E":"E"]

#pull out Row plus specific columns as DF
df3["E":"E"][["V2", "V4"]]
```
